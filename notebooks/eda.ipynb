```python
import json
import pandas as pd

data = []

with open("C:/Users/shiva/OneDrive/Desktop/Projects/AutoJudge/data/problems_data.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)
df.shape
```

```python
import numpy as np
df.replace("",np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
import sys
sys.path.append('../src')
from preprocess import clean_text

df['full_text'] = (
    df['title'] + " " +
    df['description'] + " " +
    df['input_description'] + " " +
    df['output_description']
)

df['full_text'] = df['full_text'].apply(clean_text)
```

```python
import pandas as pd
df = df.dropna()
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words="english"
)

X_tfidf = tfidf.fit_transform(df["full_text"])
```

```python
X_tfidf.shape
```

```python
df["text_length"] = df["full_text"].apply(len)
```

```python
def count_math_symbols(text):
    return sum(text.count(sym) for sym in ["+", "-", "*", "/", "<", ">", "="])

df["math_symbols"] = df["full_text"].apply(count_math_symbols)
```

```python
from scipy.sparse import hstack

X_final = hstack([
    X_tfidf,
    df[["text_length", "math_symbols"]].values
])
```
